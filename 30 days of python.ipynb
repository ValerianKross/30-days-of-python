```python
#7.Quadratic equation is calculated as follows: ax² + bx + c = 0. Write a function which calculates solution set of a quadratic equation, solve_quadratic_eqn.def solve_quadratic_eqn(a, b, c):
def solve_quadratic_eqn(a, b, c):
    discriminant = b ** 2 - 4 * a * c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2 * a)
        root2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return root1, root2
    elif discriminant == 0:
        root = -b / (2 * a)
        return root
    else:
        return 'No Real Solutions'

# Test Function
print(solve_quadratic_eqn(1, -3, 2))  

```

```python
#3. Write different functions which take lists. They should calculate_mean, calculate_median, calculate_mode, calculate_range, calculate_variance, calculate_std (standard deviation).
from collections import Counter
import math

def calculate_statistics(lst):
    # Mean
    mean = sum(lst) / len(lst)
    # Median
    lst_sorted = sorted(lst)
    n = len(lst)
    mid = n // 2
    median = lst_sorted[mid] if n % 2 != 0 else (lst_sorted[mid - 1] + lst_sorted[mid]) / 2
    # Mode
    freq = Counter(lst)
    max_count = max(freq.values())
    mode = [k for k, v in freq.items() if v == max_count]
    # Range
    range_value = max(lst) - min(lst)
    # Variance
    variance = sum((x - mean) ** 2 for x in lst) / len(lst)
    # Standard Deviation
    std_deviation = math.sqrt(variance)
    return {
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Range': range_value,
        'Variance': variance,
        'Standard Deviation': std_deviation
    }
numbers = [1, 2, 2, 3, 4, 4, 4, 5, 6]
result = calculate_statistics(numbers)
print(result)
```

```python
# Chain two or more list iterators (eg. arr.map(callback).filter(callback).reduce(callback))
from functools import reduce

# Given list of words
words = ["hello", "John", "life", "is", "a", "precious", "gift"]

# Chaining map, filter, and reduce
sentence = reduce(
    lambda acc, word: acc + " " + word,  # Concatenate words
    filter(
        lambda word: len(word) > 2,  # Keep words longer than 2 characters
        map(lambda word: word.capitalize(), words)  # Capitalize each word
    )
)

# Print the final sentence
print(sentence)
```

```python
# Write a pattern which identifies if a string is a valid python variable
import re
import keyword

def is_valid_variable(var_name):
    pattern = r'^[a-zA-Z_][a-zA-Z0-9_]*$'  # Regex pattern for valid variable names
    return bool(re.match(pattern, var_name)) and var_name not in keyword.kwlist

# Test cases
print(is_valid_variable('first_name'))  # True
print(is_valid_variable('first-name'))  # False
print(is_valid_variable('1first_name'))  # False
print(is_valid_variable('firstname'))  # True
print(is_valid_variable('def'))  # False (Python keyword)
print(is_valid_variable('_variable'))  # True
```

```python
# Read this url and find the 10 most frequent words. romeo_and_juliet
# Import the requests library for downloading the text from the URL.
import requests
# Import the re module to use regular expressions for word extraction.
import re
# Import Counter from the collections module to count word frequencies.
from collections import Counter

def top_10_frequent_words(url):
    """
    Downloads text from a given URL, processes it to extract words, and returns the 10 most common words.
    
    Parameters:
        url (str): The URL of the text file.
        
    Returns:
        list of tuples: A list of (word, count) tuples for the 10 most frequent words.
    """
    # Download the text from the URL using an HTTP GET request.
    response = requests.get(url)
    # Store the text content from the HTTP response.
    text = response.text

    # Use a regular expression to extract words:
    # \b\w+\b matches sequences of word characters (letters, digits, underscore) surrounded by word boundaries.
    # text.lower() converts the text to lowercase for case-insensitive counting.
    words = re.findall(r'\b\w+\b', text.lower())
    
    # Count the frequency of each word using Counter.
    word_counts = Counter(words)
    
    # Get and return the 10 most common words as a list of (word, count) tuples.
    return word_counts.most_common(10)

# This block runs if the script is executed directly.
if __name__ == "__main__":
    # URL for Romeo and Juliet text from Project Gutenberg.
    romeo_and_juliet = 'http://www.gutenberg.org/files/1112/1112.txt'
    # Call the function to get the top 10 most frequent words.
    top_words = top_10_frequent_words(romeo_and_juliet)
    # Print a header message.
    print("The 10 most frequent words in Romeo and Juliet are:")
    # Loop through the list of top words and print each word with its count.
    for word, count in top_words:
        print(f"{word}: {count}")
```

```python
#  find:the min, max, mean, median, standard deviation of cats' weight in metric units.
import requests             # To send HTTP requests to the API.
import statistics           # To compute mean, median, and standard deviation.
from collections import defaultdict  # To build a frequency table for country and breeds.

def parse_range(range_str):
    """
    Parses a range string (e.g., "3 - 5" or "12 - 15") and returns a list of floats.
    
    Parameters:
        range_str (str): The string representing a range of values.
    
    Returns:
        list of float: The numbers extracted from the range string.
    """
    # Split the string by '-' and remove extra spaces.
    parts = range_str.split('-')
    return [float(part.strip()) for part in parts]

def average_of_range(range_str):
    """
    Calculates the average value from a range string.
    
    Parameters:
        range_str (str): A string representing a range (e.g., "3 - 5").
    
    Returns:
        float: The average of the numbers in the range.
    """
    values = parse_range(range_str)
    return sum(values) / len(values)

def compute_stats(values):
    """
    Given a list of numerical values, computes statistical metrics.
    
    Parameters:
        values (list of float): The values to analyze.
    
    Returns:
        dict: A dictionary with min, max, mean, median, and standard deviation.
    """
    return {
        "min": min(values),
        "max": max(values),
        "mean": statistics.mean(values),
        "median": statistics.median(values),
        "std_dev": statistics.stdev(values) if len(values) > 1 else 0
    }

def main():
    # URL for The Cat API breeds endpoint.
    cats_api = 'https://api.thecatapi.com/v1/breeds'
    # Send a GET request to the API.
    response = requests.get(cats_api)
    # Parse the JSON response into a Python list.
    breeds = response.json()
    
    # Lists to store the computed average weight and lifespan for each breed.
    weights = []
    lifespans = []
    # A dictionary to store breeds by their country of origin.
    country_breed_freq = defaultdict(list)
    
    # Loop through each breed in the API response.
    for breed in breeds:
        # Retrieve the weight in metric units; e.g., "3 - 5"
        weight_str = breed.get('weight', {}).get('metric')
        if weight_str:
            # Calculate the average weight for the breed.
            avg_weight = average_of_range(weight_str)
            weights.append(avg_weight)
        
        # Retrieve the lifespan string; e.g., "12 - 15"
        lifespan_str = breed.get('life_span')
        if lifespan_str:
            # Calculate the average lifespan for the breed.
            avg_life = average_of_range(lifespan_str)
            lifespans.append(avg_life)
        
        # Retrieve the country of origin.
        origin = breed.get('origin', 'Unknown')
        # Retrieve the breed's name.
        name = breed.get('name', 'Unknown')
        # Append the breed name to the corresponding country list.
        country_breed_freq[origin].append(name)
    
    # Compute statistics for weights and lifespans.
    weight_stats = compute_stats(weights)
    lifespan_stats = compute_stats(lifespans)
    
    # Print weight statistics.
    print("Cat Weight Statistics (in kg, averaged per breed):")
    for stat, value in weight_stats.items():
        print(f"{stat}: {value:.2f}")
    
    # Print lifespan statistics.
    print("\nCat Lifespan Statistics (in years, averaged per breed):")
    for stat, value in lifespan_stats.items():
        print(f"{stat}: {value:.2f}")
    
    # Create and print a frequency table of cat breeds by country.
    print("\nFrequency Table of Cat Breeds by Country of Origin:")
    for country, breed_list in sorted(country_breed_freq.items(), key=lambda x: x[0]):
        print(f"{country}: {len(breed_list)} breeds")
        # Uncomment the following line to also print the breed names:
        # print("  " + ", ".join(breed_list))

# This block ensures the script runs only when executed directly.
if __name__ == "__main__":
    main()
```

```python

import requests                           # To send HTTP requests to the API.
from collections import Counter           # To count language frequencies.

def get_countries_data(api_url="https://restcountries.com/v2/all"):
    """
    Fetches country data from the REST Countries API.
    
    Parameters:
        api_url (str): The URL of the REST Countries API endpoint.
        
    Returns:
        list: A list of country data dictionaries.
    """
    # Send an HTTP GET request to the API.
    response = requests.get(api_url)
    # If the request is successful, return the parsed JSON data.
    if response.status_code == 200:
        return response.json()
    else:
        # Print an error message if the request failed.
        print("Error: Could not retrieve data")
        return []

def largest_countries(countries, n=10):
    """
    Finds the n largest countries by area.
    
    Parameters:
        countries (list): The list of country data dictionaries.
        n (int): The number of top countries to return.
        
    Returns:
        list: The top n countries sorted by area in descending order.
    """
    # Sort the countries by the 'area' key in descending order.
    # If a country doesn't have an 'area' key, default to 0.
    sorted_countries = sorted(countries, key=lambda x: x.get("area", 0), reverse=True)
    return sorted_countries[:n]

def most_spoken_languages(countries, n=10):
    """
    Finds the most spoken languages across all countries.
    
    Parameters:
        countries (list): The list of country data dictionaries.
        n (int): The number of top languages to return.
        
    Returns:
        tuple: A tuple containing:
            - A list of (language, count) tuples for the top n languages.
            - An integer representing the total number of unique languages.
    """
    language_counter = Counter()  # Create a Counter to tally languages.
    for country in countries:
        # Each country has a 'languages' key that is a list of language dictionaries.
        languages = country.get("languages", [])
        for lang in languages:
            # Each language dictionary should have a 'name' key.
            language_name = lang.get("name")
            if language_name:
                language_counter[language_name] += 1
    # Return the n most common languages and the total count of unique languages.
    return language_counter.most_common(n), len(language_counter)

def main():
    # Get all country data from the API.
    countries = get_countries_data()
    if not countries:
        return  # Exit if no data was retrieved.
    
    # 1. Find the 10 largest countries by area.
    top_10_largest = largest_countries(countries, 10)
    print("10 Largest Countries by Area:")
    for country in top_10_largest:
        name = country.get("name")                # Get the country's name.
        area = country.get("area", "N/A")           # Get the country's area.
        print(f"{name}: {area}")
    
    # 2. Find the 10 most spoken languages and the total number of unique languages.
    top_10_languages, total_languages = most_spoken_languages(countries, 10)
    print("\n10 Most Spoken Languages:")
    for language, count in top_10_languages:
        print(f"{language}: {count}")
    
    # 3. Print the total number of unique languages.
    print(f"\nTotal number of unique languages in the countries API: {total_languages}")

# Ensure the script runs only when executed directly.
if __name__ == "__main__":
    main()
```

```python
import requests  # To fetch the webpage
from bs4 import BeautifulSoup  # To parse HTML content

def get_uci_datasets(url="https://archive.ics.uci.edu/ml/datasets.php"):
    """
    Scrapes dataset names and their links from the UCI Machine Learning Repository.
    
    Parameters:
        url (str): The UCI dataset webpage URL.
    
    Returns:
        list: A list of tuples containing dataset names and their respective URLs.
    """
    # Send a request to fetch the HTML content
    response = requests.get(url)
    
    # Check if request was successful
    if response.status_code != 200:
        print("Error: Failed to retrieve data.")
        return []

    # Parse the webpage content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all dataset links (they are within <table> elements)
    datasets = []
    
    # UCI dataset links are inside table rows with <a> tags
    for link in soup.find_all('a', href=True):
        href = link['href']
        if "datasets/" in href:  # Check if it is a dataset link
            dataset_name = link.text.strip()  # Extract dataset name
            dataset_url = "https://archive.ics.uci.edu/ml/" + href  # Full URL
            datasets.append((dataset_name, dataset_url))
    
    return datasets

def main():
    # Get UCI datasets
    uci_datasets = get_uci_datasets()
    
    # Print the first 10 datasets as a sample
    print("Sample UCI Machine Learning Datasets:")
    for i, (name, url) in enumerate(uci_datasets[:10], start=1):
        print(f"{i}. {name}: {url}")

if __name__ == "__main__":
    main()
```

```python
from collections import Counter  # Import Counter from collections module to count occurrences in data
import math  # Import math module for mathematical operations

class Statistics:  # Define a class named Statistics
    def __init__(self, data):  # Constructor method to initialize data
        self.data = data  # Store the input data as an instance variable
    
    def count(self):  # Method to count the number of elements in data
        return len(self.data)  # Return the length of the data list
    
    def sum(self):  # Method to calculate the sum of elements in data
        return sum(self.data)  # Return the sum of all elements in the data list
    
    def min(self):  # Method to find the minimum value in data
        return min(self.data)  # Return the smallest value in the data list
    
    def max(self):  # Method to find the maximum value in data
        return max(self.data)  # Return the largest value in the data list
    
    def range(self):  # Method to calculate the range of data
        return self.max() - self.min()  # Return the difference between max and min values
    
    def mean(self):  # Method to calculate the mean (average) of data
        return self.sum() / self.count()  # Return sum divided by count
    
    def median(self):  # Method to calculate the median of data
        sorted_data = sorted(self.data)  # Sort the data list
        n = self.count()  # Get the count of elements
        mid = n // 2  # Find the middle index
        if n % 2 == 0:  # Check if count is even
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2  # Return average of middle elements
        else:
            return sorted_data[mid]  # Return the middle element if count is odd
    
    def mode(self):  # Method to find the mode of data
        freq = Counter(self.data)  # Count occurrences of each element
        max_count = max(freq.values())  # Find the highest occurrence count
        modes = [key for key, value in freq.items() if value == max_count]  # Get elements with max count
        return {'mode': modes[0], 'count': max_count}  # Return mode and its count
    
    def var(self):  # Method to calculate variance of data
        mean_value = self.mean()  # Get the mean of data
        return sum((x - mean_value) ** 2 for x in self.data) / (self.count() - 1)  # Return variance formula
    
    def std(self):  # Method to calculate standard deviation of data
        return math.sqrt(self.var())  # Return square root of variance
    
    def freq_dist(self):  # Method to calculate frequency distribution
        freq = Counter(self.data)  # Count occurrences of each element
        return sorted([(count / self.count() * 100, value) for value, count in freq.items()], reverse=True)  # Return sorted frequency distribution

ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]  # List of ages

data = Statistics(ages)  # Create an instance of Statistics with the sample data

print('Count:', data.count())  # Print count of elements in data
print('Sum: ', data.sum())  # Print sum of all elements in data
print('Min: ', data.min())  # Print the minimum value in data
print('Max: ', data.max())  # Print the maximum value in data
print('Range: ', data.range())  # Print the range of data
print('Mean: ', data.mean())  # Print the mean of data
print('Median: ', data.median())  # Print the median of data
print('Mode: ', data.mode())  # Print the mode and its count
print('Standard Deviation: ', round(data.std(), 1))  # Print the standard deviation rounded to 1 decimal place
print('Variance: ', round(data.var(), 1))  # Print the variance rounded to 1 decimal place
print('Frequency Distribution: ', data.freq_dist())  # Print the frequency distribution of data
```

```python
class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname  # Store first name
        self.lastname = lastname  # Store last name
        self.incomes = []  # List to store income entries (description, amount)
        self.expenses = []  # List to store expense entries (description, amount)
    
    def add_income(self, description, amount):
        self.incomes.append((description, amount))  # Add new income to list
    
    def add_expense(self, description, amount):
        self.expenses.append((description, amount))  # Add new expense to list
    
    def total_income(self):
        return sum(amount for _, amount in self.incomes)  # Sum of all incomes
    
    def total_expense(self):
        return sum(amount for _, amount in self.expenses)  # Sum of all expenses
    
    def account_balance(self):
        return self.total_income() - self.total_expense()  # Balance calculation
    
    def account_info(self):
        return (f"Account Holder: {self.firstname} {self.lastname}\n"
                f"Total Income: {self.total_income()}\n"
                f"Total Expense: {self.total_expense()}\n"
                f"Balance: {self.account_balance()}")  # Return formatted account details

person = PersonAccount("Maker", "Panom")
person.add_income("Salary", 5000)
person.add_income("Freelance", 1500)
person.add_expense("Rent", 1200)
person.add_expense("Groceries", 300)

print(person.account_info())  # Print account details
```

```python
import requests
from bs4 import BeautifulSoup
import json

# Wikipedia URL for the U.S. Presidents list
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States"

try:
    response = requests.get(url, timeout=10)
    response.raise_for_status()
except requests.RequestException as e:
    print(f"Error fetching the webpage: {e}")
    exit()

soup = BeautifulSoup(response.content, "html.parser")

# Locate the main table containing U.S. Presidents
table = soup.find("table", class_="wikitable")

if not table:
    print("Error: Could not find the presidents' table. The page structure might have changed.")
    exit()

presidents = []

# Extract table rows
rows = table.find_all("tr")[1:]  # Skip the header row

for row in rows:
    cols = row.find_all(["td", "th"])
    
    if len(cols) < 5:  # Ensure row has enough columns
        continue
    
    # Extract relevant details
    number = cols[0].text.strip()  # President number
    name = cols[1].text.strip().split("\n")[0]  # President name (handle links and extra text)
    term = cols[2].text.strip()  # Term duration
    party = cols[3].text.strip()  # Political party
    vice_president = cols[4].text.strip()  # Vice President(s)

    presidents.append({
        "Number": number,
        "Name": name,
        "Term": term,
        "Party": party,
        "Vice President(s)": vice_president
    })

# Save data to JSON file
json_filename = "us_presidents.json"
with open(json_filename, "w", encoding="utf-8") as json_file:
    json.dump(presidents, json_file, indent=4)

print(f"✅ Data successfully extracted and saved to '{json_filename}'")
```
